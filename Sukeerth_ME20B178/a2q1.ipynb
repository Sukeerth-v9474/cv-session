```python
import cv2
# from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt

img = cv2.imread('media/einstein.jpg')
plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)

blur_gaus_5 = cv2.GaussianBlur(img,(5,5),1.4,1.4)
blur_gaus_9 = cv2.GaussianBlur(img,(9,9),1.4,1.4)
edge = blur_gaus_5 - blur_gaus_9;

plt.figure(),plt.axis("off"),plt.title("Gaussian Blur 5x5"),plt.imshow(blur_gaus_5)
plt.figure(),plt.axis("off"),plt.title("Gaussian Blur 9x9"),plt.imshow(blur_gaus_9)
plt.figure(),plt.axis("off"),plt.title("Edges of the picture"),plt.imshow(edge)
plt.show()
```
